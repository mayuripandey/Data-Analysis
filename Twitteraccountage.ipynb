Time series Analysis

```python
import pandas as pd
df = pd.read_csv('onlyaGE.csv')
```

```python
df.info()
```

```python
df_time = df.copy()
```

```python
# set the timestamp column as the index and delete the column
df_time.index = df_time['author.created_at']
del df_time['author.created_at']
```

```python
df_time.head()
```

```python
df['date'] = pd.DatetimeIndex(df['author.created_at']).date
df['month']= pd.DatetimeIndex(df['author.created_at']).month
df['Weekday']=pd.DatetimeIndex(df['author.created_at']).weekday
df['Weekdayname']=pd.DatetimeIndex(df['author.created_at']).day_name()
df['count'] = 1
df_filtered = df[['date', 'count','month','Weekday','Weekdayname']]
df_filtered.head(5)
```

```python
df
```

```python
df = df.drop('author.created_at', 1)
```

```python
df
```

```python
df_time = df.copy()
```

```python
df_time.index = df_time['date']
```

```python
del df_time['date']
```

```python
df_time.head()
```

```python
df = df.sort_values(by="date")
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df = df.set_index('date') 
timestamp = (df!=0).resample('AS').sum()
```

```python
df
```

```python
print('Total twitter account in 2012:', df['2012']['author.id'].count())
```

```python
print('Total twitter account in 2022:', df['2022']['author.id'].count())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# plot some of the above data
# define a list of columns to plot
plot_cols = ['author.id']

# set the figure size
plt.figure(figsize=(15,8))
 
# create a color palette
palette = plt.get_cmap('Set1')


    
# now re do the most important curve, but with bigger line width
plt.plot((df!=0)['author.id'].resample('AS').sum(), marker='', color='slateblue', linewidth=3)

# add a legend with custom labels
mylabels = ['count']
plt.legend(labels = mylabels)

# add labels and title
plt.xlabel('Year')
plt.ylabel('Number of people')
plt.title('Account age by year');
```
